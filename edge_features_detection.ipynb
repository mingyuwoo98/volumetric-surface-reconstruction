```python
import matplotlib.image as image
import matplotlib.pyplot as plt
from skimage import feature
from skimage.color import rgb2gray
```

```python
sample_im = image.imread("dinoSparseRing/dinoSR0004.png")
plt.imshow(sample_im)
```

```python
# Edge detection

imgray = rgb2gray(sample_im)
can = feature.canny(imgray, 2.0)
plt.imshow(can,cmap="gray")
```

```python
# Conner detection 
coords = feature.corner_peaks(feature.corner_harris(imgray), min_distance=5, threshold_rel=0.002)
coords_subpix = feature.corner_subpix(imgray, coords, window_size=10)

fig, ax = plt.subplots()
ax.imshow(sample_im, cmap=plt.cm.gray)
# ax.plot(coords[:, 1], coords[:, 0], color='cyan', marker='o',
#         linestyle='None', markersize=6)
ax.plot(coords_subpix[:, 1], coords_subpix[:, 0], '+r', markersize=15)
plt.show()
```
